```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import KNNImputer
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import (classification_report, confusion_matrix, roc_auc_score, 
                             roc_curve, auc, accuracy_score, f1_score)
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
from imblearn.over_sampling import SMOTE
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
import lime
import lime.lime_tabular
import warnings
import tensorflow as tf
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.inspection import permutation_importance

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
warnings.filterwarnings('ignore')

# Function to load and explore data
def load_and_explore_data(file_path):
    print("\n=== Loading and Exploring Data ===")
    df = pd.read_csv(file_path)
    print("Dataset Head:")
    print(df.head())
    print("\nDataset Info:")
    print(df.info())
    print("\nDataset Description:")
    print(df.describe())
    return df

# Function for preprocessing and feature engineering
def preprocess_data(df):
    print("\n=== Preprocessing Data ===")
    initial_rows = len(df)
    df.dropna(subset=['stroke'], inplace=True)
    dropped_rows = initial_rows - len(df)
    print(f"Dropped {dropped_rows} rows due to missing target values")
    
    imputer = KNNImputer(n_neighbors=3)
    df['bmi'] = imputer.fit_transform(df[['bmi']])
    
    print("Performing feature engineering...")
    df['bmi_category'] = pd.cut(df['bmi'], bins=[0, 18.5, 24.9, 29.9, 100], 
                               labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
    df['hypertension_heart_disease'] = df['hypertension'] * df['heart_disease']
    df['age_glucose_interaction'] = df['age'] * df['avg_glucose_level']
    smoking_map = {'never smoked': 0, 'formerly smoked': 1, 'smokes': 2, 'Unknown': 1}
    df['smoking_risk_score'] = df['smoking_status'].map(smoking_map)
    work_map = {'children': 0, 'Never_worked': 0, 'Govt_job': 1, 'Private': 2, 'Self-employed': 3}
    df['work_stress'] = df['work_type'].map(work_map)
    
    nominal_cols = ['gender', 'ever_married', 'Residence_type', 'bmi_category']
    numerical_cols = ['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi', 
                     'smoking_risk_score', 'work_stress', 'hypertension_heart_disease', 
                     'age_glucose_interaction']
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('nom', OneHotEncoder(drop='first', sparse_output=False), nominal_cols),
            ('num', 'passthrough', numerical_cols)
        ])
    
    print("Applying preprocessing...")
    features = preprocessor.fit_transform(df)
    feature_names = (preprocessor.named_transformers_['nom'].get_feature_names_out(nominal_cols).tolist() + 
                     numerical_cols)
    readable_feature_names = [name.replace('_', ' ').title() if not name.startswith(('gender_', 'ever_married_', 
                                                                                   'Residence_type_', 'bmi_category_')) 
                             else (name.replace('gender_', '').replace('ever_married_', 'Married' if 'Yes' in name else 'Not Married')
                                   .replace('Residence_type_', '').replace('bmi_category_', '') + ' ' + 
                                   ('Gender' if 'gender_' in name else 'Residence' if 'Residence_type_' in name else 'BMI'))
                             for name in feature_names]
    
    labels = df['stroke']
    selector = SelectKBest(score_func=f_classif, k=min(10, features.shape[1]))
    features_selected = selector.fit_transform(features, labels)
    selected_indices = selector.get_support(indices=True)
    selected_features = [readable_feature_names[i] for i in selected_indices]
    print(f"Selected Features: {selected_features}")
    
    return features_selected, labels, selected_features

# Function to balance, split, and scale data
def balance_and_split_data(features, labels):
    print("\n=== Balancing, Splitting, and Scaling Data ===")
    smote = SMOTE(random_state=42)
    features_resampled, labels_resampled = smote.fit_resample(features, labels)
    print(f"After SMOTE: {len(labels_resampled)} samples")
    
    X_train, X_test, y_train, y_test = train_test_split(features_resampled, labels_resampled, test_size=0.2, random_state=42)
    print(f"Training samples: {len(X_train)}, Test samples: {len(X_test)}")
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    return X_train_scaled, X_test_scaled, y_train, y_test

# Visualization function
def visualize_data(df, X_scaled, y, feature_names):
    print("\n=== Generating Visualizations ===")
    numeric_df = df.select_dtypes(include=['number'])
    plt.figure(figsize=(12, 8))
    sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm', annot_kws={"size": 10})
    plt.title("Correlation Matrix of Numeric Features")
    plt.tight_layout()
    plt.show()
    
    sns.pairplot(df[['age', 'bmi', 'avg_glucose_level', 'stroke']], hue='stroke', diag_kind='kde')
    plt.suptitle("Pair Plot of Key Features", y=1.02)
    plt.show()
    
    plt.figure(figsize=(8, 6))
    ax = sns.countplot(x='stroke', data=df)
    total = len(df)
    for p in ax.patches:
        percentage = f'{100 * p.get_height() / total:.1f}%'
        ax.annotate(percentage, (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
    plt.title("Class Distribution with Percentages")
    plt.xlabel("Stroke (0: No, 1: Yes)")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# Function to evaluate a model
def evaluate_model(y_test, y_pred, y_proba, model_name):
    print(f"\n=== Evaluation for {model_name} ===")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    
    plt.figure(figsize=(8, 6))
    cm = confusion_matrix(y_test, y_pred, normalize='true')
    sns.heatmap(cm, annot=True, fmt=".2f", cmap='Blues')
    plt.title(f"Normalized Confusion Matrix - {model_name}")
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()
    
    plt.figure(figsize=(8, 6))
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'ROC AUC: {roc_auc:.2f}')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve - {model_name}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    return {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': classification_report(y_test, y_pred, output_dict=True)['1']['precision'],
        'recall': classification_report(y_test, y_pred, output_dict=True)['1']['recall'],
        'f1': f1_score(y_test, y_pred),
        'roc_auc': roc_auc_score(y_test, y_proba)
    }

# Function to explain model predictions
def explain_model(model, X_train_scaled, X_test_scaled, y_train, feature_names, model_name):
    print(f"\n=== Explainability Analysis for {model_name} ===")
    try:
        model.fit(X_train_scaled, y_train)
        X_sample = X_test_scaled[:50]
        y_pred = model.predict(X_test_scaled)
        y_proba = model.predict_proba(X_test_scaled)[:, 1]
        
        # Global Feature Importance
        print(f"Generating Feature Importance for {model_name}...")
        if hasattr(model, 'feature_importances_'):
            importance = model.feature_importances_
        else:
            perm_importance = permutation_importance(model, X_test_scaled, y_pred, n_repeats=5, random_state=42)
            importance = perm_importance.importances_mean
        feature_importance = pd.DataFrame({
            'Feature': feature_names,
            'Importance': importance
        }).sort_values(by='Importance', ascending=False)
        print(f"Feature Importance for {model_name}:")
        print(feature_importance.round(4))
        
        plt.figure(figsize=(12, 6))
        sns.barplot(x='Importance', y='Feature', data=feature_importance.head(10))
        plt.title(f"Top 10 Feature Importance - {model_name}")
        plt.xlabel("Importance")
        plt.tight_layout()
        plt.show()
        
        # LIME Local Explanation
        print(f"Generating LIME Explanation for {model_name}...")
        lime_explainer = lime.lime_tabular.LimeTabularExplainer(
            X_train_scaled,
            feature_names=feature_names,
            class_names=['No Stroke', 'Stroke'],
            mode='classification'
        )
        instance = X_sample[0]
        exp = lime_explainer.explain_instance(instance, model.predict_proba, num_features=5)
        lime_values = exp.as_list()
        feature, weight = zip(*lime_values)
        plt.figure(figsize=(10, 6))
        plt.barh(feature, weight, color=['blue' if w < 0 else 'orange' for w in weight])
        plt.title(f"LIME Local Explanation for Instance - {model_name}")
        plt.xlabel("Weight")
        plt.tight_layout()
        plt.show()
        
    except Exception as e:
        print(f"Error in explainability analysis for {model_name}: {str(e)}")
        print("Skipping explainability plots for this model...")
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.text(0.5, 0.5, f"Explainability Failed for {model_name}\nError: {str(e)}", ha='center', va='center', fontsize=12, color='red')
        ax.axis('off')
        plt.title(f"Placeholder - {model_name}")
        plt.tight_layout()
        plt.show()
        plt.close(fig)

# Function to train and compare traditional models
def train_and_compare_models(X_train_scaled, X_test_scaled, y_train, y_test, feature_names):
    print("\n=== Training and Comparing Models ===")
    models = {
        'Logistic Regression': LogisticRegression(random_state=42),
        'Random Forest': RandomForestClassifier(random_state=42),
        'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
        'LightGBM': LGBMClassifier(random_state=42),
        'SVM': SVC(probability=True, random_state=42),
        'Naive Bayes': GaussianNB(),
        'KNN': KNeighborsClassifier(),
        'Decision Tree': DecisionTreeClassifier(random_state=42),
        'Gradient Boosting': GradientBoostingClassifier(random_state=42),
        'AdaBoost': AdaBoostClassifier(random_state=42)
    }
    
    print("Tuning XGBoost hyperparameters...")
    param_grid = {'n_estimators': [100, 200], 'max_depth': [3, 5], 'learning_rate': [0.01, 0.1]}
    grid_search = GridSearchCV(XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42), 
                              param_grid, cv=3, scoring='f1', n_jobs=-1)
    grid_search.fit(X_train_scaled, y_train)
    models['XGBoost'] = grid_search.best_estimator_
    print(f"Best XGBoost Parameters: {grid_search.best_params_}")
    
    results = {}
    for name, model in models.items():
        print(f"\nTraining {name}...")
        try:
            model.fit(X_train_scaled, y_train)
            y_pred = model.predict(X_test_scaled)
            y_proba = model.predict_proba(X_test_scaled)[:, 1]
            results[name] = evaluate_model(y_test, y_pred, y_proba, name)
            explain_model(model, X_train_scaled, X_test_scaled, y_train, feature_names, name)
        except Exception as e:
            print(f"Error training {name}: {str(e)}")
            continue
    
    return models, results

# Function to train deep learning model
def train_deep_model(X_train_scaled, X_test_scaled, y_train, y_test):
    print("\n=== Training Deep Learning Model (MLP) ===")
    mlp = Sequential([
        Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    mlp.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    early_stop = EarlyStopping(patience=5, restore_best_weights=True)
    mlp.fit(X_train_scaled, y_train, validation_split=0.1, epochs=50, callbacks=[early_stop], batch_size=32, verbose=0)
    
    y_pred = (mlp.predict(X_test_scaled) > 0.5).astype('int32').flatten()
    y_proba = mlp.predict(X_test_scaled).flatten()
    results = evaluate_model(y_test, y_pred, y_proba, "MLP")
    
    return mlp, results

# Function to train stacking ensemble
def train_stacking_ensemble(models, X_train_scaled, X_test_scaled, y_train, y_test):
    print("\n=== Training Stacking Ensemble ===")
    estimators = [('rf', models['Random Forest']), ('xgb', models['XGBoost']), ('lgbm', models['LightGBM'])]
    stacking_model = StackingClassifier(estimators=estimators, final_estimator=RandomForestClassifier(random_state=42))
    stacking_model.fit(X_train_scaled, y_train)
    y_pred = stacking_model.predict(X_test_scaled)
    y_proba = stacking_model.predict_proba(X_test_scaled)[:, 1]
    results = evaluate_model(y_test, y_pred, y_proba, "Stacking Ensemble")
    
    return stacking_model, results

# Function to compare all models
def compare_all_models(traditional_results, deep_results, ensemble_results):
    print("\n=== Comparing All Models ===")
    combined = {**traditional_results, **deep_results, **ensemble_results}
    metrics_df = pd.DataFrame(combined).T[['accuracy', 'precision', 'recall', 'f1', 'roc_auc']]
    
    plt.figure(figsize=(14, 8))
    metrics_df.plot(kind='bar', colormap='tab20', width=0.8)
    plt.title('Model Performance Comparison')
    plt.ylabel('Score')
    plt.xlabel('Model')
    plt.xticks(rotation=45, fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(fontsize=10)
    plt.tight_layout()
    plt.show()
    
    leaderboard = metrics_df.sort_values(by='f1', ascending=False)
    print("\nFinal Model Leaderboard (Sorted by F1-Score):")
    print(leaderboard.round(4))

# Main execution
def main():
    print("\n=== Starting Brain Stroke Detection Pipeline ===")
    df = load_and_explore_data("brain_stroke.csv")
    features, labels, feature_names = preprocess_data(df)
    visualize_data(df, features, labels, feature_names)
    X_train_scaled, X_test_scaled, y_train, y_test = balance_and_split_data(features, labels)
    models, traditional_results = train_and_compare_models(X_train_scaled, X_test_scaled, y_train, y_test, feature_names)
    mlp, deep_results = train_deep_model(X_train_scaled, X_test_scaled, y_train, y_test)
    stacking_model, ensemble_results = train_stacking_ensemble(models, X_train_scaled, X_test_scaled, y_train, y_test)
    compare_all_models(traditional_results, {'MLP': deep_results}, {'Stacking Ensemble': ensemble_results})
    print("\n=== Pipeline Completed ===")

if __name__ == "__main__":
    main()
```
